```python
import pandas as pd
```

```python
#Load the dataset
df= pd.read_csv(r"C:\Users\user\OneDrive\Data_Science\Internship\Task1\Titanic-Dataset.csv")
```

```python
df.head()
```

```python
df.tail()
```

```python
#Fill missing 'Age' values with the median
df['Age'] = df['Age'].fillna(df['Age'].median())
```

```python
#Fill missing 'Embarked' values with the mode 
df['Embarked']=df['Embarked'].fillna(df['Embarked'].mode()[0])
```

```python
#drop the 'Cabin' column due to too many missing values
df.drop('Cabin', axis=1,inplace=True)
```

```python
# Check if there are any missing values in each column
df.isnull().sum()
```

```python
#Remove duplicates
df.drop_duplicates(inplace=True)
```

```python
#Handle outliers
import matplotlib.pyplot as plt
import seaborn as sns

# Box plot to visualize the distribution of 'Age' and identify outliers
sns.boxplot(x=df['Age'])
plt.title('Age Distribution')
plt.show()
```

```python
#Visualizations
###Bar chart for survival
sns.countplot(x='Survived' ,data=df)
plt.title('Survival Count')
plt.show()
```

```python
#Histogram for age distribution
df['Age'].hist(bins=20)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
#Correlation Heatmap for numeric features
# Select only numeric columns for the correlation heatmap
numeric_df = df.select_dtypes(include='number')

# Plot a heatmap of correlations
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

# Insights from Titanic Dataset

## Survival Rate
- Women had a higher survival rate than men.
- The survival rate was significantly higher for *first-class* passengers compared to *second* and *third-class*.

## Age Distribution Insight
- Most passengers were between *20-40 years old*, with very few passengers older than 60 years.
- The distribution of age is skewed towards younger passengers.

## Correlation Insights
- There is a strong *negative correlation* between *Pclass* and *Fare. This means that **first-class* passengers generally paid higher fares compared to *third-class* passengers.
- The *Age* and *Survived* variables have a slight *negative correlation*, meaning younger passengers had a slightly higher chanceofsurvival.
